## 1 Abertura de mapa

## 1.1 sobre o shapefile
O Esri Shapefile ou simplesmente shapefile é um formato popular de arquivo contendo dados geoespaciais em forma de vetor usado por  Sistemas de Informações Geográficas também conhecidos como  SIG.  Foi desenvolvido e regulamentado por Esri como uma especificação aberta para interoperabilidade por dados entre os softwares de Esri e de outros fornecedores.

Shapefiles espacial descrevem geometrias:  pontos,  linhas, e  polígonos. Entre outras coisas, essas geometrias podem representar Poços, Rios, e Lagos, respectivamente. Cada item pode ter atributos que os descrevem, por exemplo: nome, temperatura ou profundidade. $\textit{(Wikipédia)}$

```r
require("rgdal") 
require("ggplot2")
require("ggmap")
require("plyr")
require("sp")
```

## 3.1 Método convencional 

No método convencional, você visualiza o mapa usando a função convencional **plot**,
para baixar o mapa utilizamos os comandos:
* para abrir o arquivo shapefile
    - ogrListLayes()
    - readOGR()
* para mudar as coordenadas de **WGS84** para **Longitude\Latitude**
    - spTransform()
* finalmente plotar e botar um título
    - plot()
    - title()

```r
setwd(dir = "./mapa/");
file_map <- "mun_lin.shp";
x<- ogrListLayers(file_map);
sp=readOGR(file_map, layer=x)  
sp <-  spTransform(sp, CRS("+proj=longlat +datum=WGS84"));
plot(sp);
title("Mapa de São Paulo")
```

## 3.2 Mapa no ggplot

Dado os comandos anteriores de abertura do shapefile, agora convertemos para o modelo ser apresentado pelo *ggplot2*

```r

sp@data$id = rownames(sp@data)
sp.points  = fortify(sp, region="id")
sp.df      = join(sp.points, sp@data, by="id")
```

criamos uma nova tabela chamada **sp.df** para plotar com ggplot2.

No código abaixo temos:
* ggplot(): inicializa o objeto ggplot
* geom_polygon()
    - data: é a tabela de dados 
    - aes : recebe os parâmetros a serem plotados
          - x,y: recebe o vetor das coordenadas geográficas longitude/latitude
          - group: agrupa cada polígono pela coluna group
    - fill: pinta cada polígono do mapa com uma cor
    - colour: pinta a borda de cada polígono com uma cor
    - opcional:
        - size: valor numérico em [0,1] que define a largura da linha que cobre o polígono
        - alpha:  valor numérico em [0,1] que define a transparência da cor do polígono
* theme_bw(): define o fundo branco do gráfico
* coord_fixed(): ajusta o mapa no gráfico

```r
 ggplot()+ 
 geom_polygon(data=sp.df, aes(long,lat,group = group  ),fill="white",colour="black")+
 ggtitle("Mapa de São Paulo")+
 theme_bw() +
 coord_fixed()
```

## 3.3 Mapa com imagem de satélite

Para plotarmos o gráfico sobre o mapa de **satélite** da região de São Paulo usando o código abaixo,
onde bbox são os limites do mapa, em que você quer a área de satélite :


```r
sp@bbox[,1]<-sp@bbox[,1] - 2
sp@bbox[,2]<-sp@bbox[,2] + 2
```

* get_map(): faz a requisição da imagem de satélite para o google

```r

map <- get_map(c(sp@bbox),maptype="satellite")
ggmap(map)+
geom_polygon(aes(x = long, y = lat, group=group), alpha= 0,colour="white", fill="white", size = 0.5, data = sp.df)  
```

## 4 Processo de krigagem

## 4.1 sobre a krigagem




<a href="https://cran.r-project.org/web/packages/geoR/geoR.pdf">geoR</a> pacote para utlização de krigagem nos dados


<a href="https://cran.r-project.org/web/packages/dplyr/dplyr.pdf">dplyr</a> utiliza funções de banco de dados, no nosso caso left

```r
require("geoR")
require("dplyr")
```

carregamos a tabela de dados de série temporal pré calculada:
* codigo: código dos pluviômetros da base de dados 
* ano: anos da série temporal, [2013-2015]
* mes: variável numérica de 0 a 12 representando os meses
* latitude e longitude: coordenadas dos pluviômetros
* mm.sum: volume em milímetros de chuva total no período de um mês

```r
serie_temp <- read.csv2("../dados/serie_temporal_mensal.csv")
head(serie_temp)
meses <- c("janeiro","fevereiro","marco","abril","maio","junho",
           "julho","agosto","setembro","outubro","novembro","dezembro")
```

Faremos agora a Krigagem para o mês de **abril** de **2015**

```r
ano <- 2014
mes <- 4
titulo <- paste(mes,"de",ano,sep=" ")

```

Criamos a série nova (**serie**) com apenas os dados referentes ao ano 2015 e mês 1 (**abril**) e em seguida selecionamos as colunas (**latitude,longitude,mm.sum**)

```r

serie <-serie_temp[serie_temp$ano == ano & serie_temp$mes == mes , ]
variaveis <- c("latitude","longitude","mm.sum")
serie <- serie[,variaveis]
head(serie)
```

Converte a tabela **serie** no formato para a krigagem, no caso geodata recebe os indices das colunas longitude/latitude e mm.sum

```r
geodata <- as.geodata(serie,coords.col = c(1,2),data.col = 3)
```

Define os limites dos pontos dos pluviômetros com uma margem de 0.1

```r
xrange <- c(min(geodata[[1]][,1])-0.1,max(geodata[[1]][,1])+0.1)
yrange <- c(min(geodata[[1]][,2])-0.1,max(geodata[[1]][,2])+0.1)
```

Cria a malha com os limites dos pluviômetros com a distância de 0.05 entre eles

```r
grid <- expand.grid(seq(xrange[1],xrange[2],by=.05),seq(yrange[1],yrange[2],by=.05))
```

Calcula o variograma omnidirecional assumindo que a média do erro (u) é:
* 1st : polinômio de 1º grau mu(x) = beta0 + beta1*x1 + beta2*x2.
* 2nd : polinômio de 2º grau μ(x)= beta0 + beta1*x1 + beta2*x2 + beta3*(x1)^2 + beta4*(x2)^2 + beta5*x1*x2.
* cte : é constante 

```r
variogram <- variog(geodata, trend="2nd", uvec=seq(0, .5, by=0.05))
plot(variogram)
```

Criamps uma tabela com valores iniciais para estimar a distribuição de milímetros de chuva na área do mapa 

```r
initial.values <- expand.grid(seq(min(variogram$v), max(variogram$v), by=10),
                              seq(min(variogram$u), max(variogram$u),by=0.005))
```

selecionamos o modelo no qual queremos estimar a distribuição de chuva na região (**gaussiano**)

os modelos usando a função variofit() podem ser:
* cauchy 
* gencauchy
* circular
* cubic
* gaussian
* exponential
* matern
* spherical
* power
* powered.exponential 
* gneiting
* gneiting.matern
* wave
* pure.nugget 

que podem ser explicados nesse <a href="http://www.inside-r.org/packages/cran/geor/docs/cov.spatial">link</a> 

os parâmetros da função variofit, que estima a distribuição dado o modelo que queremos usar para estimar o semivariograma:
* fix.nugget: recebe operador lógico [TRUE/FALSE], caso FALSE o efeito pepita (**nugget**) é estimado
* nugget: recebe o valor do efeito pepita caso fix.nugget seja igual a TRUE
* ini.cov.pars: recebe a matriz de valores iniciais criadas anteriormente

```r
model <- "cubic"
fit <- variofit(variogram, cov.model= model , ini.cov.pars=initial.values,
                    fix.nugget=FALSE)
    plot(variogram)
    title(paste(meses[mes],"de",ano,"-", "modelo", model ))
    lines(fit)
 
```

Fazendo isso para todos os modelos temos:

```r
grid <- expand.grid(seq(xrange[1],xrange[2],by=.05),seq(yrange[1],yrange[2],by=.05))



variogram <- variog(geodata, trend="2nd", uvec=seq(0, .5, by=0.05))
plot(variogram)

initial.values <- expand.grid(seq(min(variogram$v), max(variogram$v), by=10),
                              seq(min(variogram$u), max(variogram$u),by=0.005))


models<-c("matern", "exponential", "gaussian",
          "spherical", "circular", "cubic", "wave")
for(model in models){
  fit <- variofit(variogram, cov.model= model , ini.cov.pars=initial.values,
                  fix.nugget=FALSE)
  plot(variogram)
  title(paste(meses[mes],"de",ano,"-", "modelo", model ))
  lines(fit)
  
}
```

agora construímos o grid com o modelo que mais se aproxima da distribuição do semi-varograma, no exemplo usamos o **gussiano**:
* krige.control:
    - type.krige: podemos utilizar vários tipos de krigagem:
        - *ok*: krigagem normal (ordinária)
        - *sk*: krigagem simples
    - cov.pars: parâmetros estimados pelo modelo
    - nugget  : efeito pepita da amostra, no caso 

```r
model <- "cubic"
krigagem <- krige.control(type.krige = "ok", cov.model = model,
                          cov.pars = fit$cov.pars,nugget = min(variogram$v))
krigagem2 <- krige.conv(geodata, locations = grid, krige = krigagem)
```

contour(), plota o mapa de calor:
    * data: dados de krigagem 
    * filled: recebe operador lógico TRUE/FALSE 
    * coords.data: localiza os pontos da amostra
    * col: passa um vetor de cores para preencher o mapa. (no caso utilizei 25 cores da função terrain.colors)
    * zlim: é a faixa de valores preditos, no caso pego o menor e o maior preditos

```r
contour(krigagem2,
        filled = TRUE,coords.data = geodata$coords,zlim = c(min(krigagem2$predict),max(krigagem2$predict)))
contour(krigagem2,
        filled = TRUE ,zlim = c(min(krigagem2$predict),max(krigagem2$predict)))
```

cria uma tabela da **malha** das coordenadas e dos valores estimados pela krigagem

```r
malha <- data.frame(grid[,1],grid[,2],krigagem2$predict)
write.csv2(malha,paste("malha","_",model,".csv",sep = ""),row.names = FALSE)
dir()
```

Criei a função a seguir para, recebendo um vetor de valores x, devolvemos um vetor de carácteres com faixas referentes aos valores de entrada

```r
cria_faixa<- function (x){
  # x <- seq(0,4000,50)
  faixas <- c(seq(0,250,50),500,1000,2000,Inf)
  ref <-data.frame(matrix(ncol = 2,nrow = length(faixas)))
  names(ref) <-  c("brks","brks2")
  for(i in 1:(length(faixas)-1)){
    if(i + 1 <10  ){
      ref[i,2] <- paste(0,0,i,"-","[",faixas[i],"-",faixas[i+1],"]",sep="") 
    }
    else{
      ref[i,2] <- paste(0,i,"-","[",faixas[i],"-",faixas[i+1],"]",sep="")
    }
    ref[i,1]<-paste(faixas[i],"-",faixas[i+1])  
  }
  
  faixa_nova <- data.frame(matrix(ncol = 1,nrow = length(x)))
  names(faixa_nova) <-  c("brks")
  for(j in 1:length(x)){
    for(i in 1:(length(faixas)-1)){
      if (faixas[i] <= x[j] && x[j] <= faixas[i+1]) faixa_nova[j,1] <- paste(faixas[i],"-",faixas[i+1])  
    }
  } 
  require("dplyr")
  z<- dplyr::left_join(faixa_nova,ref,by="brks",match=all,type="full")
  z[1:length(x),2]
  
}
```

Exemplo da função **cria_faixa()** para um vetor de dados:

```r
X = c(50,150,750,1000,1500)
data.frame(X,cria_faixa(X))
```

 Agora carregamos a malha criada anteriormente e renomeanos as colunas (**longitude**,**latitude**,**valor_predito**) para **x**,**y**,**z** e criamos uma coluna nova com a faixa daquele valor usando a função *cria_faixa()*

```r
model = "cubic"
paste("malha","_",model,".csv",sep = "")
arquivos <-paste("malha","_",model,".csv",sep = "")
gg<- read.csv2(file = arquivos )
names(gg)<- c("y","x","z")
gg$brks <- cria_faixa(gg$z)
modelo <- gsub(x=arquivos,pattern = ".csv","")

```

cria uma tabela com todas as faixas para aparecer no **ggplot**

```r
z <-c(seq(0,250,50),500,1000,2000,Inf)
 
x <- seq(min(geodata[[1]][,1])-0.5,min(geodata[[1]][,1])-0.4,length(z))
y <- seq(min(geodata[[1]][,2])-0.5,min(geodata[[1]][,2])-0.4,length(z))

test <- data.frame(y=y ,x= x ,z= z, brks = cria_faixa(z))
gg<-rbind(gg,test)
```

Tabela com cores pré escolhidas para ser usada para cada faixa

```r
    color  <-c("#a50026","#d73027",  "#f46d43",  "#fdae61",  "#fee08b",
               "#d9ef8b",  "#a6d96a",  "#66bd63",  "#1a9850",  "#006837")
    color <- color[order(1:10,decreasing = TRUE)]
```

agora plotamos a malha nos mapas ggplot e satélite:

```r
 ggplot() %+% gg+aes(x,y)+
      geom_polygon(data=sp.df, aes(long,lat,group=group,fill=""),fill="white",colour="black") +
      geom_tile(aes(fill=brks,alpha=0.6 ))+
      scale_fill_manual("mm",values= color)+
      ggtitle(paste("SP-",modelo,ano,meses[mes],sep=" ")) 
```

E plotamos sobre a imagem de satélite

```r
 ggmap(map)  %+% gg+aes(x,y)+
      geom_polygon(data=sp.df, aes(long,lat,group=group ),fill = "white" , colour="white",alpha=0,size= 0.2) +
      geom_tile(aes(fill=brks,alpha=0.6 ))+
      scale_fill_manual("mm",values= color)+
      ggtitle(paste("SP_ -",modelo,ano,meses[mes],sep=" ")) +
      coord_equal()
```
